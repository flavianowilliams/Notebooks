```python
from m3l.structure import System as sys
import json
#
system = sys()
system.loadSystem('system_Ar.json')
system.atoms
```

```python
x = [item[1].item() for item in system.atoms]
y = [item[2].item() for item in system.atoms]
z = [item[3].item() for item in system.atoms]
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot(xs = [x[1], x[2]], ys = [y[1], y[2]], zs = [z[1], z[2]], color ='r')
ax.scatter(xs = x, ys = y, zs = z)
ax.set_xlabel('x')
ax.set_ylabel('y')
#ax.set_zlabel('z')
#ax.view_init(90, -90, 0)
#ax.axis('off')
```
